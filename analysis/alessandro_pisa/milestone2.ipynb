# Conclusion #
---
Overall there are many differnt characteristics to mushrooms and edible and poisonous mushrooms share many of them. Because of this there is no particular formula to guarantee that a mushroom is edible or not, nonetheless there are enough differences for us to create a hypothesis with a high likelyhood of acertion. The following are things one should consider when judging a mushroom (and both of them should be considered):

**1. Edible Mushrooms Could Have These Characteristics:**
    - Have no particular odor, or smell like anise or almond (big teller).
    - Have a gray or red rooted stalk (altough these are not very common they are likely edible).
    - Have orange or brown veil (once again not very common, but very likely edible).
    - Pendant ring (does not guarantee edibility).
    - Brown Sprore Print (does not guarantee edibility).
    - Grows on waste habitat and has an abundant/numerous population. (Edible mushrooms also grow in many other places but so do poisonous ones).
    
    
**2. Poisonous Mushrooms Could Have These Characteristics:**
    - Odor, anything that smells bad, spicy, fishy, in general foul (very good teller).
    - Chocolate, or white spore print colors.
    - Grow on paths (keep in mind they can grow in many other places, but path mushrooms tend to be poisonous).
    - A large singular ring, or no ring on the mushroom.
    - A silky stalk colored either cinammon, yellow, or buff (both above and below the ring).
    - A buff colored gill (another good indicator as if they have this they are very likeley poisonous).
    
Keep in mind these are just guidelines, and there is nothing stopping a perfectly edible mushroom to have characterisitcs of a poisonous one or vice versa. Overall these gudilines can help us form a better judgement of the edibility of a mushroom, but it is always better to play it safe if we really don't know.
