# Segmenting the Data #
---
Since there are so many distinct characteristics to a mushroom we have opted to separate these into different categories for an easier and more organized analysis. These include: cap, gill, stalk, viel, ring, life and miscellaneous traits.

### How to Read the Data ###
---
We have two distinct types of bargraphs below, the first one displays the *count* of edible and poisonous mushrooms with a certain characteristic (this are the blue and light blue graphs), directly below them you can find the associated "Edibility Ratio" bargraphs in which we calculate the ratio of the amount of edible mushrooms compared to the total amount of mushrooms with that specific trait, the formula for the ratio is: 

\\(e / t\\) 

Where e is the number of *edible* mushrooms with a specific trait, and t is *total* number of mushrooms with that trait.

Overall when it comes to the Edibility Ratio bargraph the closer the value is to 1, the better that characteristic serves as an indicator for the mushrooms edibility.

### Cap Related Traits ###
---
Overall when it comes to the Cap of the mushroom there is really not enough prevelant distinctions to determine if a mushroom is poisonous or not. Many of the features are shared between both edible and poisonous mushrooms, for example: a convex cap shape, a scaly surface or a brown or gray mushroom cap. Nontheless, there are some characteristics — like a knobbed cap shape and red, yellow or buff colored caps — that seem to be a lot more common between poisonous mushrooms. There is one trait, grooves in the cap surface, that are distinclty traits of a poisonous mushroom, but these are very rare. 

```python
project_functions.show_cap_related_features(df, ratioDict)
```

### Gill Related Traits ###
---
When it comes to the gill of the mushroom there are actually a couple traits than can help us distingush between an edible or poisonous mushroom. The most prevelant one is the gill color, as it appears that green and buff colored gill's are really good indicators for the toxicity of a mushroom. Similarly, if a mushroom has a narrow gill it would be better to assume it as toxic as there is more than double the amount of poisonous mushrooms with narrow gills than edible. As for the other characteristics if a mushroom has crowded gill spacing or atached gills it is likely for it to be edible, but not definite.

```python
project_functions.show_gill_related_features(df, ratioDict)
```

### Stalk Related Traits ###
---
For the stalk of the mushroom there seems to be a relation with the roughness of the stalk both above and below the ring as well as with the color of the stalk. Overall mushrooms with a silky stalk colored either cinammon, yellow, or buff (both above and below the ring) appear to be overall poisonous. In this case the better indicator is the color, as if the stalk happens to be of one of those colors it is most definitley poisonous. It important to mention thataltough they are good indicators, these traits tend to be more rare as they form a very small percentage out of the entire sample. 

Another notable characteristic is that mushrooms with a rooted stalk are very likely to be edible, as well as mushrooms with gray or red stalks (both above and below ring).

```python
project_functions.show_stalk_related_features(df, ratioDict)
```

### Veil Related Traits ###
---
In the case of the veil, mushrooms with an orange or brown veil seem to be edible, altough they are quite rare. When it comes to a partial veil both poisonous and edible mushrooms share this characteristic.

### Ring Related Traits ###

As for the ring of the mushrooms, it appears that mushrooms with no rings, or large rings are poisonous. This appears to be a good indicator for the toxicity.
